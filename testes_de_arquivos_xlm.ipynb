```python
import cv2
from google.colab import drive
drive.mount('/content/drive')
```

```python
imagem = cv2.imread('/content/drive/MyDrive/Visão Computacional Guia Completo-20220604T072031Z-001/Visão Computacional Guia Completo/Images/people1.jpg')
from google.colab.patches import cv2_imshow

cv2_imshow(imagem)
```

```python
detectar_pessoas = cv2.CascadeClassifier('/content/drive/MyDrive/Visão Computacional Guia Completo-20220604T072031Z-001/Visão Computacional Guia Completo/Cascades/haarcascade_frontalface_default.xml')
imagem2 = cv2.imread('/content/drive/MyDrive/Visão Computacional Guia Completo-20220604T072031Z-001/Visão Computacional Guia Completo/Images/people2.jpg')
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
detectar = detectar_pessoas.detectMultiScale(imagem2, scaleFactor=1.3, minNeighbors=3, minSize=(32,32))

for (x, y, w, h) in detectar:
    cv2.rectangle(imagem2, (x,y), (x+w, y+h), (180, 0, 100), 2)
cv2_imshow(imagem2)
```

```python
detectar_carros = cv2.CascadeClassifier('/content/drive/MyDrive/Visão Computacional Guia Completo-20220604T072031Z-001/Visão Computacional Guia Completo/Cascades/cars.xml')
carros = cv2.imread('/content/drive/MyDrive/carros pra teste.jpg')
detector_carros = detectar_carros.detectMultiScale(carros, scaleFactor=1.05, minNeighbors=2, minSize=(60,60))

for (x, y, w, h) in detector_carros:
     cv2.rectangle(carros, (x, y),(x+w, y+h), (255, 0, 0), 3)

cv2_imshow(carros)
print(detector_carros)
```
